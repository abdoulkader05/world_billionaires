```python
# import the librairies

import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
billionaires = pd.read_csv('./Billionaire.csv')
billionaires.head(10)Top 5 domains with the most numbers of billionaires 
```

```python
# preprocess the data

billionaires.isnull().sum()
billionaires = billionaires.dropna()
```

```python
billionaires["NetWorth"] = billionaires["NetWorth"].astype(str).str.strip("$")
billionaires["NetWorth"] = billionaires["NetWorth"].str.strip("B")
billionaires["NetWorth"] = billionaires["NetWorth"].astype(float)
```

```python
billionaires.head()
```

## Analyze according to their Net worth

```python
top = 10

bill = billionaires.sort_values(by=["NetWorth"], ascending=False).head(top)

plt.figure(figsize=(20,10))

sns.barplot(bill['Name'], bill['NetWorth'])
```

## Top 5 domains with the most numbers of billionaires 

```python
source = billionaires['Source'].value_counts().head(5)
keys = source.values
id = source.index
palette_color = sns.color_palette('bright')
plt.figure(figsize=(30,10))
plt.pie(keys, labels = id)
plt.title("Top 5 with the most number of billionaires ")
plt.show()
```

```python
source = billionaires['Industry'].value_counts().head(5)
keys = source.values
id = source.index
palette_color = sns.color_palette('bright')
plt.figure(figsize=(30,10))
plt.pie(keys, labels = id)
plt.title("Top 5 with the most number of billionaires ")
plt.show()
```

```python
source = billionaires['Country'].value_counts().tail(20)
keys = source.values
id = source.index
palette_color = sns.color_palette('bright')
plt.figure(figsize=(40,10))
plt.pie(keys, labels = id,  autopct='%.0f%%')
plt.title("Top 5 with the most number of billionaires ")
plt.show()
```
